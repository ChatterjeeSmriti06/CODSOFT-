```python
numbers = {}
emails= {}
choice = 0
while choice != 5:
    choice = int(input(" 1. Print Phone Numbers \n 2. Add a Phone Number \n 3. Remove a Phone Number \n 4. Find a Phone Number \n 5. Quit \n Type a number : "))
    if choice == 1:
        print("Telephone Numbers:")
        for x in numbers.keys():
            print("Name: ", x, "\t Number:", numbers[x],"\t Email:",emails[x])
        print()
    elif choice == 2:
        print("Add Name,Number and Email1")
        name = input("Name: ")
        phone = input("Number: ")
        email= input("Email")
        numbers[name] = phone
        emails[name]= email
    elif choice == 3:
        print("Remove Name, Number and Email")
        name = input("Name: ")
        if name in numbers:
            del numbers[name]
        else:
            print(name, "was not found")
    elif choice == 4:
        print("Find a Number")
        name = input("Name: ")
        if name in numbers:
            print("The number is", numbers[name])
        else:
            print(name, "was not found")
    elif choice != 5:
        print("Enter a valid choice")
```
